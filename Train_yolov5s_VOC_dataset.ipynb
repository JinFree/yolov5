커스텀 데이터셋을 훈련하는 경우의 훈련 절차는 아래와 같습니다.

1. LabelImg 등을 이용하여 YOLO 형태(txt)의 어노테이션 파일을 준비할 것
2. 훈련용 데이터 폴더와 검증용 데이터 폴더에 이미지 파일과 같은 이름의 txt 어노테이션 파일을 위치시킬 것
3. yaml 파일을 작성할 것

----

yaml 파일의 기본 구조

train: train 데이터셋의 폴더 경로 또는 이미지 파일 리스트가 적힌 텍스트 파일 경로

val: validation 데이터셋의 폴더 경로 또는 이미지 파일 리스트가 적힌 텍스트 파일 경로

nc: 검출 대상의 클래스 수

names: 검출 대상 클래스 이름, 리스트 형식 ex) names: ['car', 'person']

----
